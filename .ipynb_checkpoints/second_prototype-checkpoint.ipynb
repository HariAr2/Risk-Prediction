```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
class NeuralNetwork(nn.Module):
    def __init__ (self,input_size,hidden_size,num_classes):
        super(NeuralNetwork,self).__init__()
        self.l1 = nn.Linear(input_size,hidden_size)
        self.relu1 = nn.ReLU6()
        self.bn1 = nn.BatchNorm1d(hidden_size)
        self.sig1 = nn.Sigmoid()
        self.relu2 = nn.ReLU6()
        self.sig2 = nn.Sigmoid()
        self.relu3 = nn.ReLU6()
        self.sig3 = nn.Sequential()
        self.relu4 = nn.ReLU6()
        self.l2 = nn.Linear(hidden_size,num_classes)
    def forward(self,x):
        x = self.l1(x)
        x = self.relu1(x)
        x = self.bn1(x)
        x = self.sig1(x) 
        x = self.relu2(x)
        x = self.sig2(x)
        x = self.relu3(x)
        x = self.sig3(x)
        x = self.relu4(x)
        x = self.l2(x)
        return x
```

```python
df = pd.read_csv('/Users/hari/Downloads/Maternal Health Risk Data Set.csv.xls')
df
```

```python
risk_level_map = {'high risk':2,'mid risk':1,'low risk':0}
df['RiskLevel'] = df['RiskLevel'].map(risk_level_map)
```

```python
df
```

```python
train_data , test_data = train_test_split(df,random_state=42,shuffle=True,test_size=0.3)
```

```python
train_data
```

```python
test_data
```

```python
X_train = train_data.drop(columns=['RiskLevel'])
y_train = train_data['RiskLevel']
```

```python
X_test = test_data.drop(columns=['RiskLevel'])
y_test = test_data['RiskLevel']
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train = torch.tensor(X_train,dtype=torch.float32)
X_test = torch.tensor(X_test,dtype=torch.float32)
```

```python
y_train = torch.tensor(np.array(y_train),dtype=torch.long)
y_test = torch.tensor(np.array(y_test),dtype=torch.long)
```

```python
input_size = X_train.shape[1]
hidden_size = 1024
num_classes = len(np.unique(y_train))
```

```python
model = NeuralNetwork(input_size,hidden_size,num_classes)
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
```

```python
num_epochs = 8000
for epoch in range (num_epochs):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs,y_train)
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 4000 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    test_outputs = model(X_test)
    _, predicted = torch.max(test_outputs, 1)
    accuracy = (predicted == y_test).sum().item() / y_test.size(0)
    print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score

with torch.no_grad():
    test_outputs = model(X_test)  
    _, predicted = torch.max(test_outputs, 1)  

y_true = y_test.cpu().numpy()   
y_pred = predicted.cpu().numpy() 


precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

print(classification_report(y_true, y_pred))
```

# MODEL BUILD

As we can see from the graphs most of the parameters  are important in finding the risk level in the below we will first divide the data as TEST,TRAIN and VAL and use different algorithms to find the best and use it 

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
train_data, temp = train_test_split(data,random_state=42,test_size=0.2)
```

```python
val_data,test_data = train_test_split(temp,random_state=42,test_size=0.5)
```

```python
train_data
```

```python
val_data
```

```python
test_data
```

```python
X = train_data.drop(columns=['RiskLevel'])
y = train_data['RiskLevel']
```

```python
model = DecisionTreeRegressor(random_state=42)
model.fit(X,y)
```

```python
X_test = test_data.drop(columns=['RiskLevel'])
y_test = test_data['RiskLevel']
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test,y_pred)
r2_score = r2_score(y_test,y_pred)

print(f"Mean Squared Error MSE: {mse}")
print(f"R2 score: {r2_score}")
```

## Graphical Representation of Graph

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=X.columns, filled=True, rounded=True)
plt.show()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_val = val_data.drop(columns=['RiskLevel'])
y_val = val_data['RiskLevel']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_val = scaler.fit_transform(X_val)
X_test = scaler.fit_transform(X_test)
```

```python
log_reg = LogisticRegression(multi_class='multinomial',
                             random_state=42,
                             solver='lbfgs',
                             max_iter=1000)
log_reg.fit(X,y)

y_val_pred = log_reg.predict(X_val)
accuracy = accuracy_score(y_val,y_val_pred)
classification_rep = classification_report(y_val,y_val_pred)
print(f"accuracy: {accuracy}")
print(f"Classification Report : {classification_rep}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.01,0.1,1,10,100],
              'solver' : ['lbfgs','saga']
}

grid_search = GridSearchCV(LogisticRegression(
                           max_iter = 1000, 
                           random_state = 42),
                           param_grid,
                           cv=5,
                           scoring='accuracy',
                          verbose=1)
grid_search.fit(X,y)
print("Best Parameters",grid_search.best_params_)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier(max_iter=1000,random_state=42,loss='log_loss')
sgd_clf.fit(X,y)


y_test_pred = sgd_clf.predict(X_test)
accuracy = accuracy_score(y_test,y_test_pred)
print(f"accuracy: {accuracy}")
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)
print("Test Accuracy:", rf.score(X_test, y_test))
```

```python
#XGboost
xgb = XGBClassifier(random_state = 42)
xgb.fit(X,y)
print("Test Accuracy:",xgb.score(X_test,y_test))
```

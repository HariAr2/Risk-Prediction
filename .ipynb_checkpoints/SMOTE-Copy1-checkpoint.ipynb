```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
class NeuralNetwork(nn.Module):
    def __init__ (self,input_size,hidden_size,num_classes):
        super(NeuralNetwork,self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_size,hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(hidden_size,hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(),
            
            nn.Linear(hidden_size,hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(),
            nn.Dropout(0.2),
            
            nn.Linear(hidden_size,hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(),
            
            nn.Linear(hidden_size,hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(hidden_size,hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.LeakyReLU(),
            
            nn.Linear(hidden_size,num_classes)
        )
        
    def forward(self,x):
        return self.layers(x)
            
```

```python
df = pd.read_csv('/Users/hari/Downloads/Maternal Health Risk Data Set.csv.xls')
df
```

```python
risk_level_map = {'high risk':2,'mid risk':1,'low risk':0}
df['RiskLevel'] = df['RiskLevel'].map(risk_level_map)
```

```python
df
```

```python
train_data , test_data = train_test_split(df,random_state=42,shuffle=True,test_size=0.2)
```

```python
train_data
```

```python
test_data
```

```python
X_train = train_data.drop(columns=['RiskLevel','BodyTemp'])
y_train = train_data['RiskLevel']
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_resampled,y_train_resampled = smote.fit_resample(X_train,y_train)
```

```python
import numpy as np
from collections import Counter

print("Before SMOTE:", Counter(y_train))
print("After SMOTE: ", Counter(y_train_resampled))
```

```python
X_test = test_data.drop(columns=['RiskLevel','BodyTemp'])
y_test = test_data['RiskLevel']
```

```python
scaler = StandardScaler()
X_train_resampled = scaler.fit_transform(X_train_resampled)
X_test = scaler.fit_transform(X_test)
```

```python
X_train_resampled = torch.tensor(X_train_resampled,dtype=torch.float32)
X_test = torch.tensor(X_test,dtype=torch.float32)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train_resampled = le.fit_transform(y_train_resampled)  
y_test = le.transform(y_test)
```

```python
import torch

y_train_resampled = torch.tensor(y_train_resampled, dtype=torch.long)
y_test = torch.tensor(y_test, dtype=torch.long)
```

```python
y_train_resampled = torch.tensor(np.array(y_train_resampled),dtype=torch.long)
y_test = torch.tensor(np.array(y_test),dtype=torch.long)
```

```python
input_size = X_train_resampled.shape[1]
hidden_size = 1024
num_classes = len(np.unique(y_train_resampled))
```

```python
model = NeuralNetwork(input_size,hidden_size,num_classes)
optimizer = torch.optim.AdamW(model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lambda epoch: 0.8 ** (epoch // 200))
```

Learning Rate = 0.001 we can change it to 0.0001 if we need to improve

```python
num_epochs = 4000
for epoch in range (num_epochs):
    optimizer.zero_grad()
    outputs = model(X_train_resampled)
    loss = criterion(outputs,y_train_resampled)
    loss.backward()
    optimizer.step()
    
    scheduler.step()
    
    if (epoch + 1) % 1000 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    test_outputs = model(X_test)
    _, predicted = torch.max(test_outputs, 1)
    accuracy = (predicted == y_test).sum().item() / y_test.size(0)
    print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score

with torch.no_grad():
    test_outputs = model(X_test)  
    _, predicted = torch.max(test_outputs, 1)  

y_true = y_test.cpu().numpy()   
y_pred = predicted.cpu().numpy() 


precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

print(classification_report(y_true, y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
with torch.no_grad():
    test_outputs = model(X_test)  
    probabilities = torch.softmax(test_outputs, dim=1)  

y_true = y_test.cpu().numpy()  
y_prob = probabilities.cpu().numpy()
```

```python
n_classes = y_prob.shape[1]  # Number of classes
y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))

fpr = {}  # False Positive Rate
tpr = {}  # True Positive Rate
roc_auc = {}  # AUC scores

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot the ROC curve
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

# Plot random classifier line
plt.plot([0, 1], [0, 1], 'k--', label="Random (AUC = 0.50)")

# Formatting the plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("AUROC Curve")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
macro_auroc = roc_auc_score(y_true_bin, y_prob, average="macro")
weighted_auroc = roc_auc_score(y_true_bin, y_prob, average="weighted")

print(f"Macro AUROC: {macro_auroc:.4f}")
print(f"Weighted AUROC: {weighted_auroc:.4f}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, classification_report

precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
auroc = roc_auc_score(y_true_bin, y_prob, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUROC: {auroc:.4f}")

print(classification_report(y_true, y_pred))
#prcision score,Recall,F1 score
```

```python
from skorch import NeuralNetClassifier

# Wrap your PyTorch model
skorch_model = NeuralNetClassifier(
    module=NeuralNetwork,
    module__input_size=input_size,
    module__hidden_size=hidden_size,
    module__num_classes=num_classes,
    max_epochs=5000,
    lr=0.001,
    optimizer=torch.optim.Adam,
    criterion=nn.CrossEntropyLoss,
    iterator_train__shuffle=True,
    verbose=0,
    device='cuda' if torch.cuda.is_available() else 'cpu'
)
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

stack = StackingClassifier(
    estimators=[
        ('xgb', xgb_model),
        ('rf', rf_model),
        ('nn', skorch_model),
    ],
    final_estimator=LogisticRegression(),
    cv=5,
    passthrough=True  # optional: includes original features too
)
```

```python
stack.fit(X_train_resampled, y_train_resampled)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = stack.predict(X_test)
y_proba = stack.predict_proba(X_test)

print(classification_report(y_test, y_pred))
print("AUROC:", roc_auc_score(y_test, y_proba, multi_class="ovr"))
```
